```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from mpl_toolkits.mplot3d import Axes3D
```

# Load the customer data from CSV file

```python
df = pd.read_csv("D:/downloads 2/customer_segmentation_data.csv")
```

# Show first few rows

```python
print(df.head())
```

# Preprocessing – Scaling

```python
scaler = StandardScaler()
scaled_data = scaler.fit_transform(df)
```

# Applying K-Means Clustering

```python

kmeans = KMeans(n_clusters=4, random_state=42)
df['Cluster'] = kmeans.fit_predict(scaled_data)
```

 # Visualize Clusters – 2D Scatter Plot

```python
plt.figure(figsize=(8,6))
sns.scatterplot(data=df, x='Income', y='Spending', hue='Cluster', palette='Set2', s=100)
plt.title("Customer Clusters (2D View)")
plt.xlabel("Annual Income")
plt.ylabel("Spending Score")
plt.grid(True)
plt.show()
```

#  Visualize Clusters – 3D Scatter Plot

```python
fig = plt.figure(figsize=(10,7))
ax = fig.add_subplot(111, projection='3d')
scatter = ax.scatter(df['Age'], df['Income'], df['Spending'],
                     c=df['Cluster'], cmap='Set1', s=80)
ax.set_title("Customer Clusters (3D View)")
ax.set_xlabel("Age")
ax.set_ylabel("Income")
ax.set_zlabel("Spending")
plt.legend(*scatter.legend_elements(), title="Cluster")
plt.show()
```
