```python
import pandas as pd
```

# Load dataset

```python
df = pd.read_csv("D:/downloads 2/survey_dataset.csv")
```

# Bar Chart â€“ Platform Preference

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(style="whitegrid")
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x='Platform Preferred', palette='Set2')
plt.title('Platform Preference')
plt.xlabel('Platform')
plt.ylabel('Number of Respondents')
plt.xticks(rotation=15)
plt.show()
```

#  Pie Chart â€“ Gender Distribution

```python
gender_counts = df['Gender'].value_counts()
plt.figure(figsize=(6, 6))
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', colors=sns.color_palette('Set3'))
plt.title('Gender Distribution')
plt.show()
```

```python
heatmap_data = pd.crosstab(df['Age Group'], df['Satisfaction (1-5)'])
plt.figure(figsize=(8, 5))
sns.heatmap(heatmap_data, annot=True, cmap='coolwarm', fmt='d')
plt.title('Satisfaction by Age Group')
plt.xlabel('Satisfaction Score')
plt.ylabel('Age Group')
plt.show()
```

```python
import plotly.express as px
import numpy as np
import pandas as pd

# If df is already defined, you might need to convert string columns
# Convert any columns with unicode data type to string
if 'df' in locals() or 'df' in globals():
    # Convert string columns to the new np.str_ type
    for col in df.select_dtypes(include=['object']).columns:
        df[col] = df[col].astype(str)

# Create the plot
fig = px.bar(df, x='Platform Preferred', color='Gender', title='Platform Preference by Gender')
fig.show()
```
